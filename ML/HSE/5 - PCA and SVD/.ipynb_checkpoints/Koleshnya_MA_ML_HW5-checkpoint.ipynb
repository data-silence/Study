### Задание 1 [2 балла] <a id="task1"></a>
Вспомнить детали реализации алгоритма главных компонент (https://arxiv.org/pdf/1404.1100.pdf) и получить оператор сжатия.

Оценки будут производиться по главным компонентам **второй** недели

```python
from scipy import linalg
from sklearn.utils.extmath import svd_flip
from sklearn.decomposition import PCA

X2_mean = X2-X2.mean(axis=0)

U, s, V = linalg.svd(X2_mean, full_matrices=False)
U, V = svd_flip(U, V)

#Попробуем подобрать оптимальное количество компонент:

sing_v = (s ** 2) / (X2_mean.shape[0] - 1)

var_ = np.round(np.cumsum([i*(1/sum(sing_v)) for i in sing_v]), decimals=3)

d = np.argmax(var_ >= 0.83) + 1

print('%.2f %% variance retained in %d dimensions' \
% (var_[d], d))
```

```python
X2_based_pc = U[:, 0:d].dot(np.diag(s)[0:d, 0:d])
print(X2_based_pc, X2_based_pc.shape)
```

```python
# воспользуемся TruncatedSVD
from sklearn.decomposition import TruncatedSVD
based_pc = TruncatedSVD().fit(X2_mean)

def compression_operator(X, mean_pc=X2_mean, based_pc=based_pc, number_pc=7):
    X = X-X.mean(axis=0)
    based_pc = TruncatedSVD(number_pc).fit(mean_pc)
    return based_pc.transform(X)

def rho(X, mean_pc=X2_mean, based_pc=based_pc, number_pc=7):
    X = X-X.mean(axis=0)
    based_pc = TruncatedSVD(number_pc).fit(mean_pc)
    X_proj = based_pc.transform(X)
    if number_pc == 1:
        explained_var = np.var(X_proj, axis=0) 
    else:
        explained_var = np.var(X_proj, axis=0)  
    total_var = np.var(X, axis=0).sum() 
    return explained_var / total_var, sum(explained_var / total_var)
```

Подсчитаем метрику $\rho := $ какую долю дисперсии данных в каждую из трех недель (по отдельности) объясняют первые $d$ главных компонент, оценненые по первой неделе. 

Пусть 
 - $X$ матрица $121 \times 2016$ для которой нужно оценить такую метрику
 - $\overline{X}$ вектор средних по строкам
 - $U$ первые $d$ главных компонент, оценненые по первой неделе
 - $Z(X) = U^T(X2 - \overline{X2})$ преобразование сжатия 
 
Тогда 

$$
\rho(X) = \dfrac{\operatorname{trace}(Z - \overline{Z})(Z - \overline{Z})^T}{\operatorname{trace}(X - \overline{X})(X - \overline{X})^T} 
$$

Выпишем преобразование восстановления для $X$:

$$\hat{X} = \overline{X2} + UZ(X)$$

где $Z(X)$ - преобразование сжатия из прошлого задания.

```python
def decompression_operator(compressed_X, mean_pc=X2.mean(axis=0), based_pc=based_pc, number_pc=7):
    based_pc = TruncatedSVD(number_pc).fit(X2_mean)
    Xinv = based_pc.inverse_transform(compressed_X)
    Xinv += mean_pc
    return Xinv  
```

```python
def residual_matrix(X): 
    return X - decompression_operator(compression_operator(X)) 
```

### Задание 4 [2 балла] <a id="task4"></a>
Теперь достаточно объявлять тревого для наблюдения с "большим" полученным остатком (ошибкой восстановления). Однако, здесь требуется уточнить, что такое "большим"?
Один из возможных подходов к определению этого уровня статистически, предполагая нормальное распределение рассчитываемой статистики, нам остается задать уровень значимости $\alpha$. Расчет статистики реализован в функции q_critical. Математические детали можно прочитать здесь: ( https://www.ltu.se/cms_fs/1.51590!/pca_part%201_jackson_1980.pdf, интересующая формула: 33).  

Вам необходимо взять параметр $\alpha$ равным 0.05 и 0.01, привести на рисунках значение попрога $Q_\alpha$ (изобразив константной линией) и рассчитать число аномалий для каждой недели.

```python
def q_critical(X, number_pc, alpha):
    X_mean = X.mean(axis=0)
    
    w, v = np.linalg.eig(np.cov((X - X_mean).T))
    l = -np.sort(-w)[number_pc:]
    
    th1 = []
    for i in l:
        th1.append(i)
    
    th2 = []
    for i in l:
        th2.append(i**2)
    
    th3 = []
    for i in l:
        th3.append(i**3)
        
    theta_one = sum(th1) 
    theta_two = sum(th2)  
    theta_three = sum(th3)
        
    h = 1 - (2*theta_one*theta_three)/(3*(theta_two**2))
    
    if h > 0:
        c_alpha = stats.norm.ppf(1 - alpha) 
    else:
        c_alpha = stats.norm.ppf(alpha) 
        
    inside_brackets =  (c_alpha*(2*theta_two*(h**2))**0.5)/theta_one + ((theta_two*h*(h - 1))/(theta_one**2)) + 1 # your code here
    return theta_one * np.power(inside_brackets, 1/h) 
 
```

```python
print('q_critical at alfa 0,01 for X2:', q_critical(X2, 7, 0.01))
print('q_critical at alfa 0,05 for X2:', q_critical(X2, 7, 0.05))
```

Добавим критические значения статистики и рассмотим график для каждой недели отдельно:

```python
def q_plot(X, X_resid, number_pc):
    f, axarr = plt.subplots(2, sharex=True, figsize=(50, 20))
    
    axarr[0].set_yscale('log')
    axarr[0].plot([q_stat(X_resid, t) for t in range(2016)])
    axarr[0].axhline(y=q_critical(X, number_pc, 0.01) , xmin=0, xmax=2016, linewidth=2, color = 'r')
    axarr[0].axhline(y=q_critical(X, number_pc, 0.05) , xmin=0, xmax=2016, linewidth=2, color = 'g')
    axarr[0].axis('tight')
    
    axarr[1].plot(X);
    axarr[1].axis('tight')
```

```python
# я решила немного поменять нижний порог для параметра гамма 
gamma = np.linspace(1e-10, 1, 10)

model = OneClassSVM(nu=nu_opt, kernel='rbf') 
grid = {
    'gamma': gamma
}

search = GridSearchCV(model, grid, scoring=scores,
                      n_jobs=-1, cv=2,
                      refit='accuracy_score')

search.fit(X_subsample, y_subsample)
```

```python
print("gamma:     ", search.cv_results_['param_gamma'].data)
print("scores: ", search.cv_results_['mean_test_accuracy_score'])
```

```python
score1 = search.cv_results_['mean_test_accuracy_score'] 
score1.max()
gamma_opt = search.cv_results_['param_gamma'].data[score1==score1.max()]
gamma_opt
```

```python
# Теперь применим найденные оптимальные ню и гамма 
from sklearn import metrics  
model = OneClassSVM(nu=nu_opt, kernel='rbf', gamma=gamma_opt) 
model.fit(X_subsample)
preds = model.predict(X_subsample)  
targs = y_subsample
print ("====== Training Dataset Accuracy=======================")
print("Training Dataset accuracy: ", metrics.accuracy_score(targs, preds))  
print("Training Dataset precision: ", metrics.precision_score(targs, preds))  
print("Training Dataset recall: ", metrics.recall_score(targs, preds))  
print("Training Dataset f1: ", metrics.f1_score(targs, preds))  
print("Training Dataset area under curve (auc): ", metrics.roc_auc_score(targs, preds))  
```
