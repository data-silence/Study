Количество записей в Машином списке покупок может быть любым. Каждая запись вводится с новой строки. Записи вводятся до тех пор, пока пользователь не введет "Конец списка". После этого одной строкой вводится буква, по которой сегодня определяется скидка. Если название продукта начинается на заданную букву, то на этот продукт действует скидка 15 процентов.

Образец вывода смотрите в примере. Если ни на один товар скидка не действует, то программа должна вывести "Скидка: 0 рублей"

*Вы считали категории в список списков, где каждая строка выглядит как в образце  - 1 балл*  
`[цена (int), название (str), количество (int)]`  
Например, `[120, мандарины (1 кг), 2]`

*Подсчет скидки - 1 балл.*

**Пример ввода**  
120 руб., мандарины (1 кг), 2 кг.  
70 руб., молоко, 3 пакета  
15 руб., батон сдобный, 1 штука  
345 руб., белое вино, 1 бутылка  
120 руб., моцарелла, 2 упаковки  
Конец списка  
м  

**Вывод**

Cкидка: 103.5 руб.


```python
shopping_list = []
dicsount = int(15)
discount_counter = 0

while True:
    try:
        l = input('Введи свои покупки через запятую в формате [цена (в рублях и без копеек), название, количество]:\n').lower()
        if l == 'конец списка':
            break
        print('\t--- Чтобы завершить ввод, набери "конец списка" ---\n')
        l_new = l.replace(' руб.', '').split(', ')
        l_new[0] = int(l_new[0])
        l_new[-1] = int((l_new[-1])[0])
        shopping_list.append(l_new)
    except:
        print('Прочти инструкцию по вводу внимательно и попробуй ещё разок: ')
        
letter = input('\nВведи букву, на которую на этой неделе в "Магнолии" скидка:\n').lower()

for el in shopping_list:
    if el[1][0] == letter:
        discount_counter += el[0] * el[-1] * (dicsount/100)
        
print(f'\nСкидка: {discount_counter} ₽')
```

# Морской бой, ver 1.0

### Известные ошибки и косяки:
- некорректно завершает программу, в случае досрочной победы
- большинство функций не включают обрабоку исключений, поэтому при неправильном вводе программа может ломаться или принимать недопустимые значения
- допускает ввод нулей в качестве пользовательских координат, что противоречит введенной системе координат
- завершающий цикл многоходовой игры очень плохо написан, но попытки реализации в более правильных циклах и логике пока порождает гораздо большее количество серьезных косяков чем упомянутый п.1 данного списка, так что пока не осилил. 

```python
"""Морской бой, версия 1.0"""


from random import randint

field = []



def set_field():
    """Определяет размерность игрового поля, подготавливает и выводит на экран"""
    global size
    size = int(input('Введите размерность игрового поля (длина стороны игрового квадрата): '))
    
    for i in range(size):
        s = []
        for j in range(size):
            s.append('O')
        field.append(s)

    for el in field:
        print(' '.join(el))


def hide_ship():
    """Генерирует размещение корабля и запоминает координаты"""
    global ship_row
    global ship_col
    
    ship_row = randint(1,5)
    ship_col = randint(1,5)


def user_guess():
    """Позволяет пользователю задать правильные координаты, и сохраняет их"""
    global guess_row
    global guess_col
    
    while True:
        try:
            user_coord = input('Как думаешь: какие координаты у вражеского корабля? Ввод нулей недопустим. Вводи через пробел:\n')
            user_coord_clean = user_coord.split(' ')
            if int(user_coord_clean[0]) > size or int(user_coord_clean[1]) > size:
                print(f'Координаты не могут превышать размерности поля, равной {size}\n')
            else:
                guess_row = int(user_coord_clean[0])
                guess_col = int(user_coord_clean[1])
                break
        except:
            print('Казалось бы, что может быть проще чем ввести два числа через пробел? Попробуй ещё раз, пожалуйста.\n')


def is_this_target(x, y):
    """Оценивает ход игрока, обрабатывает в зависимости от успешности хода"""
    if x == ship_row and y == ship_col:
        print('Ты крутой чувак, ты победил!\n')
    else:
        print(f'\nТы думал, что корабль находится в {guess_row} {guess_col}. Но это фиаско, братан!')

    field[y-1][x-1] = 'X'
    for el in field:
        print(' '.join(el))

        
def print_field():
    """Просто выводит текущее состояние игрового поля"""
    for el in field:
        print(' '.join(el))

        
"""Основной цикл программы"""

set_field()
hide_ship()

for i in range(1, 6):
    print(f'\nПопытка number {i}')
    user_guess()
    is_this_target(guess_row, guess_col)
    
    if guess_row == ship_row and guess_col == ship_col:
        break

    
print(f"\nЭто была последняя попытка, игра закончена. Корабль находился на клетке {ship_row}, {ship_col}")
field[ship_col-1][ship_row-1] = 'S'
print('Давайте посмотрим на то, как вы сыграли:')
print_field()
```
