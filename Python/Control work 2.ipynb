# Задание 2: Язык племени Мумба-юмба (2.5 балла)

Дан текст на языке племени Мумба-Юмба. Запишите в файл все слова, встречающиеся в тексте, разделяя их пробелом. Слова должны быть отсортированы по убыванию их количества появления в тексте, а при одинаковой частоте появления — в алфавитном порядке.

Словом считается последовательность непробельных символов идущих подряд, слова разделены одним или большим числом пробелов или символами конца строки. Большие и маленькие буквы считаются различными

*Подсказка.* После того, как вы создадите словарь всех слов, вам захочется отсортировать его по частоте встречаемости слова. 
Вызвав метод dict.items() от вашего словаря вы получите список кортежей вида [('what', 2), ('hi', 2), ('is', 1)]. Чтобы получилось нужная сортировка такой список нужно сначала отсортировать по алфавиту от a до z (первый элемент), а затем по убыванию их количества появления в тексте. Эту задачу можно решить через две сортировки (и не забудьте про параметр key во второй!).

*Подсказка 2.* Когда будете записывать слова в файл, не забудьте, что после последнего слова пробела быть не должно.

**Входной файл: problem2.txt**  
**Приложите к блокноту файл с ответом: output2.txt**

Первое слово в правильном ответе: qzjijwnklo, последнее: zzzsrymltasziudkfpqn

````python
import string
  
text = open("problem2.txt", "r")


'''Делаем словарь с именем d с ключами в виде слов, значениями в виде частоты их нахождения в словаре'''
d = dict()

for line in text:
    line = line.strip().lower().replace(string.punctuation, '') # нормализуем  текст
    words = line.split(" ")
    for word in words:
        if word in d:
            d[word] = d[word] + 1
        else:
            d[word] = 1
print(d)

'''Дважды сортируем'''
first_sorted_list = sorted(list(d.items()))
second_sorted_list = sorted(first_sorted_list, key = lambda x: x[1], reverse=True)
#print(second_sorted_list)

'''Делаем список только слов в нжуно порядке, который потом преобразовываем в текст требуемого формата'''
result = []
for el in second_sorted_list:
    result.append(el[0])

result_text = ' '.join(result)

'''Записываем текст в файл с требуемым названием'''
f = open("output2.txt", "w", encoding='utf8')
f.write(result_text)
f.close()
````

# Задание 3. Сумма строк (1 балл)

Дан файл, каждая строка которого может содержать одно или несколько целых чисел, разделенных одним или несколькими пробелами.

Вычислите сумму чисел в каждой строке и выведите эти суммы через пробел (для каждой строки выводится сумма чисел в этой строке).

В этой задаче вам нужно подсчитать ответ для данного входного файла. 

**Входной файл: problem3.txt**  
**Ответ выведите в блокноте. **

Первое число в ответе: 34568147, последнее: 111

Например, если бы содержимое входного файла было следующим:

```
2 2
 3    4
 1       5
```
то в качестве ответа необходимо было бы сдать последовательность чисел

4 7 6

````python
d = [] # здесь формируем список нормализованных числовых строк после извлечения строки из файла
k = [] # здесь список списков чисел: каждый внутренний список хранит числа отдельной строки 
result = [] # итоговый список, которые хранит результаты - список сумм чисел каждой строки

text = open("problem3.txt", "r")
content = text.readlines()
text.close()

"""Получаем список числовых строк с пробелом между нами"""
for line in content:
    line = ' '.join(line.strip().split())
    d.append(line)
#print(d)

'''Получаем список списков из элементов в виде числовых строк'''
for stroka in d:
    stroka = stroka.split(' ')
    k.append(stroka)
#print(k)

'''Получаем список из сумм чисел отдельных строк'''
for el in k:
    summ = 0
    for i in el:
        i = int(i)
        summ += i 
    result.append(summ)

'''Делаем из результирующего списка строку в требуемом формате: без запятых и квадратных скобок'''    
z =((str(result)[1:-1].replace(',', '')))
print(z)
````

# Задание 4: буквы, строки и слова (1.5 балла)
Дан файл. Определите сколько в нем букв (латинского алфавита), слов, строк. Выведите три найденных числа в формате, приведенном в примере. Словом считается последовательность больших и маленьких латинских букв. Все остальные символы считаются разделителями слов. Если будете решать регулярками, то будьте аккуратны, потому что некоторые выражения "захватывают" кроме букв лишние символы. Если будете пользоваться методами строк, то .isalpha() проверяет на то, что строка является латинской буквой.

Пример входного файла:
```
Beautiful is better than ugly.
Explicit is better than implicit.
Simple is better than complex.
Complex is better than complicated.
```
Пример ответа:

```

108 letters 
20 words 
4 lines 
```

**Входной файл: problem4.txt**  
**Ответ нужно вывести в блокноте по образцу в примере**

В количестве букв первая цифра 4, а последняя 3.  
В количестве слов: 1 и 6.  
В количестве строк: 1 и 0.  

````python
import string
punct = string.punctuation

def normalize(text):
    '''Нормализует текст, делая буквы прописными и убирая лишние символы. Результат помещает в список слов'''
    for el in punct:
        text = text.lower().replace(el, ' ').replace('\n', ' ')
    text_list = text.split(' ')
    norm_list = [el.strip(punct) for el in text_list if el not in punct]
    return norm_list
    
text = open("problem4.txt", "r")
content = text.read()
text.close()

norm_list = normalize(content)

string_amount = content.count('\n') # считает количество строк текста
words_amount = len(norm_list) # считает количество слов в тексте
letters_amount = len(''.join(norm_list)) # считает количество букв в тексте, убирая пробелы между словами

print('Input file contains:')
print(f'{letters_amount} letters')
print(f'{words_amount} words')
print(f'{string_amount} lines')

# Текст похож на "Дживс и Вустер" :-)
````

Входные данные:

— Первым на вход подается список с информацией о времени заезда
— Вторым — список с информацией о времени выезда
— Третьим — К, обозначающее количество номеров

Выходные данные:
— Логическое значение, которое обозначает возможность забронировать номера
false означает, что в отеле не достаточно номеров для N бронирований
true означает, что в отеле достаточно номеров для N бронирований

Пример:

Входные данные:  

* даты заезда = [1, 3, 5]
* даты выезда = [2, 6, 10]
* К = 1

Выходные данные: false. В день = 5 в отеле 2 гостя. Но у нас есть только один номер.

*Подсказка.* Один из вариантов решения может быть таким - в цикле в каждый день месяца проверять сколько комнат занято в этот день.

````python
# данные для проверки
arrivals = [1, 4, 5, 5, 8, 10, 23, 24, 24, 25]
departures = [2, 7, 8, 20, 12, 13, 27, 25, 29, 26]
rooms = 2
````

````python
# Возможно вам пригодится структура, когда даты каждого гостя объединены в кортеж (но можно решить и по-другому)
# Сделать это можно с помощью функции zip, которой передаются два списка/кортежа одинаковой длины
list(zip(['a', 'b'], [2423, 232]))
````

````python
arrivals = [1, 4, 5, 5, 8, 10, 23, 24, 24, 25]
departures = [2, 7, 8, 20, 12, 13, 27, 25, 29, 26]
rooms = 2

arrivals_tuple = tuple(arrivals)
departures_tuple = tuple(departures)
kort = list(zip(arrivals_tuple, departures_tuple)) # создаем кортеж для удобства обработки отдельного бронирования
#print(kort) 


'''Создаём список из 30 списков, каждый соответствует дню месяца. В список дня будем заносить факт брони номера'''
days = []
for day in range (30):
    start = []
    days.append(start)

'''Занесем данные о броне в каждом дне месяца в список для проведения дальнейших расчетов'''
for el in kort:
    dif = (el[1] - el[0]) # Находим количество занимаемых заказом дней
    for i in range((dif)):
        days[(el[0]+i-1)].append(1) # Отмечаем добавлением единицы в список нужных дней факт брони номера согласно заявок

print(days) # выведем итоговое представление заказов по каждому дню, с которым и будем теперь проводить нужные расчеты 
print('\f')

'''Считаем количество бронирований номеров за каждый день и выводим результат, сравнивая с номерным фондом'''
count = 0 #счетчик дней
for day in days:
    count += 1
    answer = rooms >= len(day)
    if answer == True:
        print(f'{answer}: в день #{count} номеров достаточно')
    else:
        print(f'{answer}: в день #{count} проблема: хотят забронировать {len(day)} номера, а у нас их всего {rooms}')
````
